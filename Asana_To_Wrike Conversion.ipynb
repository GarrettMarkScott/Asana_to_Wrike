```python
import pandas as pd
import random


df = pd.read_csv('Garretts_Tasks_in_Dealer_World.csv')
```

```python
df['Created At'] = pd.to_datetime(df['Created At'])
df['Created At'] = df['Created At'].dt.strftime('%m/%d/%Y')
```

```python
import datetime

#Create a datetime object with today's value
today = datetime.datetime.today() 

#print today's date in YYYY-MM-DD format
print(datetime.datetime.strftime(today,'%m/%d/%Y'))

print(today)
```

```python
#add one day to today's date
tomorrow = today + datetime.timedelta(1)
tomorrow = datetime.datetime.strftime(tomorrow,'%m/%d/%Y')
#print tomorrow's date in YYYY-MM-DD format
print(tomorrow)
```
